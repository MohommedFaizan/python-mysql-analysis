```python
import numpy as np
import pandas as pd
import calendar
```

```python
data=pd.read_csv('/content/WalmartSalesData.csv.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().any()
```

```python
data.dtypes
```

```python
data['Date']=pd.to_datetime(data['Date'] )
```

```python
data['time_of_day']=data['Time'].str[0:2]
data.head()
```

```python
data['day_of_week']=data['Date'].apply(lambda x: calendar.day_name[x.weekday()])
data.head()
```

```python
data['month_name']=data['Date'].dt.month.apply(lambda x: calendar.month_name[x])
data.head()
```

```python
data.columns
data=data.rename(columns={'Invoice ID':'Invoice_ID','Customer type':'Customer_type','Product line':'Product_line','Unit price':'Unit_price',
                          'gross margin percentage':'gross_margin_percentage','gross income':'gross_income','Tax 5%':'Tax_5%'})
```

```python
data.columns
```

```python
data.to_csv('Walmartdata.csv', index=False)
```
